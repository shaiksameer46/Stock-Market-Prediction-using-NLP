```python
def label(date):
  print(date)
  if(len(nyse.valid_days(start_date=date, end_date=date))):
    price_open = data.DataReader(name = "^GSPC", data_source = "yahoo", start = date, end = date)["Open"]
    price_close = data.DataReader(name = "^GSPC", data_source = "yahoo", start = date, end = date)["Close"]
    if(int(price_close) > int(price_open)):
      return(1)
    else:
      return(0)
  else:
    return("incorrect value")

path = "/content/drive/My Drive/NLP/chunk/"
for i in range(31,32):
  df_test = pd.read_csv(path + str(i) + ".csv")
  df_test["label"] = df_test['date'].apply(label)
  df_test.to_csv(path + str(i) + ".csv", encoding = "utf-8")
```
