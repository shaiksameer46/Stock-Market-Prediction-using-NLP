```python
nlp = spacy.load("en_core_web_sm")
```

```python
stemmer = PorterStemmer()
```

```python
df2 = pd.read_csv("/content/drive/My Drive/Colab Notebooks/NLP/labeled_data.csv")
def feature(sentence):
    words = nlp(sentence)
    words_stop = [i.lemma_ for i in words if not i.is_stop]
    words_list = [stemmer.stem(i) for i in words_stop]
    return(words_list)

df2["title"] = df2["title"].apply(feature)
```

```python
df2.astype({"label":int})
```

```python
import string 
import re 
import numpy as np
invalidcharacters= set(string.punctuation)

a = {"Text":[], "Positive": [], "Negative" : []}
df3 = pd.DataFrame(a)
print(df3["Text"].values)

def nlpframe(text, label, df3):
    print(text, label)
    for i in text:
        if any(((char in invalidcharacters) or (char.isnumeric()) or (re.search(' +', char))) for char in i):
            #invalid
            pass
        else:
          print(i)
          if(i in df3["Text"].values):
              if(label):
                  df3.loc[df3['Text'] == i, 'Positive'] += 1
              else:
                  df3.loc[df3['Text'] == i, 'Negative'] += 1
          else:
              if(label):
                  Positive = 1
                  Negative = 0
              else:
                  Negative = 1
                  Positive = 0
              Text = i
              Positive = Positive
              Negative = Negative
              df_record = {"Text": Text, "Positive" : int(Positive), "Negative" : int(Negative)}
              df3 = df3.append(df_record, ignore_index = True)
    return(df3)

for i in range(100):
    df4 = nlpframe(df2.iloc[i]["title"], df2.iloc[i]["label"], df3)
    df3 = df4
```
