```python
import pandas_market_calendars as mcal
nyse = mcal.get_calendar('NYSE')
```

```python
def filtering(date):
  if(len(nyse.valid_days(start_date=date, end_date=date))):
    return(1)
  else:
    return(0)
df["valid"] = df['date'].apply(filtering)
```

```python
df.info()
```

```python
df = df[df["valid"] == 1]
```

```python
df.drop(["valid","Unnamed: 0"], axis = 1, inplace = True)
```
